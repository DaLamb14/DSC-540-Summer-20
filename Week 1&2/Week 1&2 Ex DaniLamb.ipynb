```python
multiline_text = multiline_text.replace('\n', "")
multiline_text
```

```python
clean_text = ""
for c in multiline_text:
    if c == " ":
        clean_text += c
    elif c.isalnum():
        clean_text += c
    else: 
        clean_text += " "
clean_text
```

```python
word_list = clean_text.split()
word_list
```

```python
unique_dict = dict.fromkeys(word_list)
len(word_list), len(list(unique_dict.keys()))
```

```python
for word in word_list:
    if unique_dict[word] is None:
        unique_dict[word] = 1
    else:
        unique_dict[word] += 1
unique_dict
```

```python
top_words = sorted(unique_dict.items(), reverse = True, key = lambda key_val_tuple: key_val_tuple[1])
top_words[:25]
```

```python
def csv_parser(header, line):
    line = itertools.zip_longest(header, line, fillvalue = None)
    csv_dict = {val[0]: val[1] for val in line}
    return csv_dict
```
