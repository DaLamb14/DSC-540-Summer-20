```python
income = pd.read_csv(r"C:\Users\datre\OneDrive\Documents\Graduate School\Summer '20\adult_income_data.csv", names = text)
income.head()
```

```python
income.isnull().sum()
```

```python
income_sub = income[["age","education","occupation"]]
income_sub.head()
```

```python
plt.hist(income["age"], bins = 20)
plt.title("Histogram of Age")
plt.show()
```

```python
def strip_white(sw):
    return sw.strip()
```

```python
income_sub["edu_strip"] = income["education"].apply(strip_white)
income_sub["education"] = income_sub["edu_strip"]
income_sub.drop(labels = ["edu_strip"], axis = 1, inplace = True)
```

```python
income_sub["occ_strip"] = income["occupation"].apply(strip_white)
income_sub["occupation"] = income_sub["occ_strip"]
income_sub.drop(labels = ["occ_strip"], axis = 1, inplace = True)
```

```python
age_subset = income_sub[(income_sub["age"] >= 30) & (income_sub["age"] <= 50)]
age_subset.shape[0]
```

```python
occ_age = income_sub.groupby("occupation").describe()["age"]
occ_age
```

The category with the oldest average works are Exec-managerial at 42.2. The lowest 75th percentile is 34 which is the Armed-Forces section, however, this section only has 9 observtions so there could be doubt that this is the most accurate decision and could be an outlier. The next lowes 75th percentile is Handlers-cleaners at 39. This means that 75% of the Handlers and cleaners are below the age of 39 and 25% are above 39.

```python
sns.barplot(y = occ_age.index, x = occ_age["count"])
```

```python
df_1 = income[["occupation", "hours-per-week", "Income"]].sample(20, random_state = 101)
df_2 = income[["occupation", "education", "native-country"]].sample(20, random_state = 101)
df_merge = pd.merge(df_1, df_2, on = "occupation", how = "inner").drop_duplicates()
df_merge
```
